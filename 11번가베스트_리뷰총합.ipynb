```python
from selenium import webdriver
import time
from selenium import webdriver
import time
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import random
import pandas as pd
```

```python
# Selenium 설정
service = Service(ChromeDriverManager().install())
options = Options()
options.add_argument('--start-maximized')
options.add_argument('--disable-blink-features=AutomationControlled')

# 단일 URL 설정
url = "https://www.11st.co.kr/products/1815207391?inpu=&trTypeCd=29&trCtgrNo=895021"

# Selenium으로 크롤링 시작
driver = webdriver.Chrome(service=service, options=options)
driver.implicitly_wait(10)

try:
    driver.get(url)
    time.sleep(3)

    # 스크롤 끝까지
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    # 리뷰 iframe 전환
    iframe = driver.find_element(By.ID, "ifrmReview")
    driver.switch_to.frame(iframe)

    # "더보기" 버튼 클릭 (최대 20번)
    for _ in range(20):
        try:
            # 정확한 클래스 이름으로 "더보기" 버튼 탐지
            more_button = driver.find_element(By.CSS_SELECTOR, '.c_product_btn.c_product_btn_more8.review-next-list')
            driver.execute_script("arguments[0].click();", more_button)  # JavaScript로 클릭
            print("더보기 버튼 클릭 완료")
            time.sleep(2)  # 클릭 후 로드 대기
        except Exception as e:
            print(f"더보기 버튼이 없거나 더 이상 클릭할 수 없습니다: {e}")
            break

    # 리뷰 데이터 추출
    date_lst = []
    name_lst = []
    grade_lst = []
    rev_lst = []
    kkuk_lst = []

    date = driver.find_elements(By.CLASS_NAME, "date")
    for i in date:
        date_lst.append(i.text)

    name = driver.find_elements(By.CLASS_NAME, "name")
    for i in name:
        name_lst.append(i.text)

    grade = driver.find_elements(By.CSS_SELECTOR, "p.grade > span > em")
    for i in grade:
        grade_lst.append(i.text)

    rev = driver.find_elements(By.CLASS_NAME, "cont_review_hide.text-expanded")
    for i in rev:
        rev_lst.append((i.text).replace('\n', ''))

    kkuk = driver.find_elements(By.CLASS_NAME, "kkukCount")
    for i in kkuk:
        kkuk_lst.append((i.text))

    # DataFrame으로 변환
    data = pd.DataFrame(
        zip(date_lst, name_lst, grade_lst, rev_lst, kkuk_lst),
        columns=['날짜', 'ID', '평점', '리뷰', '꾹']
    )

    # DataFrame을 CSV 파일로 저장
    output_csv = "single_url_reviews.csv"
    data.to_csv(output_csv, index=False, encoding='utf-8-sig')
    print(f"크롤링 완료! 리뷰가 {output_csv}에 저장되었습니다.")

except Exception as e:
    print(f"오류 발생: {e}")

finally:
    driver.quit()
```
